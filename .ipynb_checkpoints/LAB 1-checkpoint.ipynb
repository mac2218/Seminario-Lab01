```python
import numpy as np
from scipy.stats import entropy

# Distribución 1
P = np.array([0.5, 0.5])

# Distribución 2
Q = np.array([0.25, 0.25, 0.25, 0.25])

H_P = entropy(P, base=2)
H_Q = entropy(Q, base=2)

print("Entropía de P:", H_P)
print("Entropía de Q:", H_Q)
```
